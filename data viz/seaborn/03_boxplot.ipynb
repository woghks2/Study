# **3. boxplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

plt.style.use('seaborn')
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Paperlogy'
```

```python
n = 300
data = {
    '국어': np.round(np.random.normal(75, 10, size=n).clip(50, 100), 1),
    '영어': np.round(np.random.normal(72, 12, size=n).clip(50, 100), 1),
    '수학': np.round(np.random.normal(70, 15, size=n).clip(50, 100), 1),
    '성별': np.random.choice(['남', '여'], size=n),
    '반': np.random.choice([f'{i}반' for i in range(1,10)], size=n)
}
 
df = pd.DataFrame(data)
```

## **1. boxplot 기본**

* x,y를 지정해주지 않으면 수치형 변수도 x축으로 이동해서 warning messege가 나올 수 있음

```python
fig, ax = plt.subplots(figsize=(8,6))
box_plot = sns.boxplot(
    data=df, x='반', y='국어')
ax.set_title('Box Plot')
plt.show()
```

## **2. example**

```python
fig, ax = plt.subplots(figsize=(10, 4))
box_plot = sns.boxplot(
    # 1. 데이터 설정 : 데이터프레임 / X축(범주) / Y축(수치)
    data=df, x='반', y='수학', ax=ax,
    # 2. 색상 설정 : 색상 구분 / 팔레트 / 채도
    hue='성별', palette='muted', saturation=0.8,
    # 3. 박스 스타일링 : 박스 너비, 테두리 두께 및 색상
    width=0.5, 
    boxprops={'linewidth': 1, 'edgecolor': 'black', 'alpha': 1},  # 박스 테두리 설정
    # 4. 중앙값 스타일링 : 색상, 두께 및 투명도
    medianprops={'color': 'black', 'linewidth': 1, 'alpha': 1},
    # 5. 수염 스타일링 : 수염 두께, 스타일, 색상 및 투명도
    whiskerprops={'linewidth': 1.5, 'linestyle': '--', 'color': 'gray', 'alpha': 0.8},
    # 6. 상자 끝 스타일링 : 상자 끝 두께, 색상 및 투명도
    capprops={'linewidth': 2, 'color': 'gray', 'alpha': 0.8},
    # 7. 이상치 스타일링 : 이상치 모양, 색상, 크기 및 투명도
    flierprops={'marker': 'o', 'markerfacecolor': 'gray', 'markersize': 5, 'alpha': 0.8}
)
plt.title('반별 수학 성적 분포', fontsize=16, pad=10)
plt.show()
```
