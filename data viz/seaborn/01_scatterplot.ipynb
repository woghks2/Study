# **1. Scatter plot**

* matplotlib==3.5.0
* seaborn==0.12

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

plt.style.use('seaborn')
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Paperlogy'
```

```python
n = 300
data = {
    '국어': np.round(np.random.normal(75, 10, size=n).clip(50, 100), 1),
    '영어': np.round(np.random.normal(72, 12, size=n).clip(50, 100), 1),
    '수학': np.round(np.random.normal(70, 15, size=n).clip(50, 100), 1),
    '성별': np.random.choice(['남','여'], size=n),
    '반': np.random.choice([f'{i}반' for i in range(1,10)], size=n)
}
 
df = pd.DataFrame(data)
```

## **1. scatter plot 기본**

```python
fig,ax = plt.subplots(figsize=(8,6))
scatter_plot = sns.scatterplot(
    data=df, ax=ax,
)
ax.set_title('Scatter Plot')
plt.show()
```

## **2. scatter**

1. **데이터 설정**
   * data=df, x='col1', y='col2',

2. **색상 설정**
   * hue='col3', alpha=0.7, palette='muted',
   * 마커 구분 색상 / 마커 투명도 / 마커 팔레트 색상

3. **마커 설정**
   * style='col4', markers=['o', 's', 'D', '^'], size='col5', sizes=(50, 200),
   * 마커 모양 분류 / 마커 모양 / 마커 크기 분류 / 마커 크기

4. **범례 설정**
   * legend='brief'
   * scatter_plot.legend(loc='center left', bbox_to_anchor=(1.0, 0.6))

```python
plt.figure(figsize=(8,6))

scatter_plot = sns.scatterplot(
    data=df, x='국어', y='영어',
    hue='성별', palette='muted', alpha=0.7,
    style='반', markers=['o', 's', 'D', '^', 'p', 'H', 'X', '8', 'P'],
    size='수학', sizes=(50, 200),
)

scatter_plot.legend(bbox_to_anchor=(1.02,1))  

plt.title("학생 성적 시각화", fontsize=16)
plt.show()
```

## **3. relplot**

1. **데이터 설정**
   * kind='scatter', data=df, x='col1', y='col2',

2. **색상 설정**
   * hue='col3', alpha=0.7, palette='muted',
   * 마커 구분 색상 / 마커 투명도 / 마커 팔레트 색상

3. **마커 설정**
   * style='col4', markers=['o', 's', 'D', '^'], size='col5', sizes=(50, 200),
   * 마커 모양 분류 / 마커 모양 / 마커 크기 분류 / 마커 크기

4. **범례 설정**
   * legend='brief'
   * scatter_plot.legend(loc='center left', bbox_to_anchor=(1.0, 0.6))
  
5. **크기 설정**
   * height=8, aspect=0.75

```python
scatter_plot = sns.relplot(
    
    # 데이터
    kind='scatter', data=df, x='국어', y='영어',
    
    # 마커 : 색상/스타일/크기
    hue='성별', palette='muted', alpha=0.7,
    style='반', markers=['o', 's', 'D', '^', 'p', 'H', 'X', '8', 'P'],
    size='수학', sizes=(50, 200),
    
    legend='brief',
    # 플롯 크기 : 높이 / 종횡비
    height=6, aspect=1.3
)

# 범례 위치 조정
scatter_plot.legend.set_bbox_to_anchor((1, 0.62))  
plt.title("학생 성적 시각화", fontsize=16)
plt.show()
```

## **4. example**

```python
fig, ax = plt.subplots(figsize=(8,6))
scatter_plot = sns.scatterplot(
    data=df, x='수학', y='영어', ax=ax,
    hue='성별', alpha=0.7, palette='muted',
    size='국어', sizes=(50, 200),
    legend='brief',
    )
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.6))
ax.set_title("학생 성적 시각화", fontsize=16, pad=20)
plt.show()
```
