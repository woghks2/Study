# **12. lineplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

plt.style.use('seaborn')
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Paperlogy'
```

```python
# 데이터 생성
np.random.seed(42)
 
# 기본 데이터 구조 생성
months = ['월', '4월', '5월', '6월', '7월']
classes = ['A반', 'B반', 'C반', 'D반']
 
data = []
for class_name in classes:
    # 기본 성적 수준 설정 (반마다 다르게)
    base_score = np.random.normal(75, 5)
    
    # 시간에 따른 변화 추가
    for month in months:
        # 약간의 상승 트렌드와 노이즈 추가
        trend = months.index(month) * 2  # 시간에 따른 상승
        noise = np.random.normal(0, 3)   # 랜덤 변동
        
        score = base_score + trend + noise
        data.append({
            '월': month,
            '반': class_name,
            '평균점수': round(score, 1)
        })
 
# DataFrame 생성
df = pd.DataFrame(data)
```

## **1. pointplot 기본**

```python
sns.lineplot(data=df, x='월', y='평균점수')
plt.show()
```

```python
sns.lineplot(data=df, x='월', y='평균점수', hue='반')
plt.show()
```

## **2. pointplot 스타일링**

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.lineplot(
    # 1. 기본 데이터 설정: 데이터프레임 / X축 / Y축 / 색상구분
    data=df, x='월', y='평균점수', ax=ax,
    # 2. 스타일 설정
    hue='반', palette='muted', alpha=0.8,
    # 3. 라인 스타일 설정
    linestyle='-', linewidth=2,
    # 4. 마커 설정
    markers=True, marker='o', markersize=8
)
ax.set_title('Line Plot', pad=15)
plt.show()
```
